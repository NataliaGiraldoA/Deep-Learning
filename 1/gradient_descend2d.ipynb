```python
import numpy as np
import matplotlib.pyplot as plt

def fx(x):
  return x**2 +0.5*np.sin(5*x)
  #return 0.1*x**4 - x**3 - 2*x**2 + 5*x
  #return np.sin(x) +0.1*x**2

def derivate(x0):
  h = 0.0000001
  return (fx(x0+h)-fx(x0))/h

def funcion(gamma, x_initial):
  x = x_initial
  all_values = [x]
  for i in range(1000):
    x = x - gamma * derivate(x)
    #print(x)
    all_values.append(x)
    if abs(derivate(x)) < 0.0000001:
      break
  return np.array(all_values)

initial_x = -0.7
gamma = 0.01

si = funcion(gamma, initial_x)

x_plot = np.linspace(-2, 2, 400)
y_plot = fx(x_plot)
plt.plot(x_plot, y_plot, 'b-', linewidth=2)
plt.plot(si,fx(si), 'ro')
plt.show()
```
